```python
from ultralytics import YOLO
from PIL import Image
import cv2

import matplotlib.pyplot as plt
Final_model = YOLO('yolo11m.yaml').load('yolo11m.pt')  

# Train
Result_Final_model = Final_model.train(data="archive/brain-tumor.yaml", epochs=10, imgsz = 640, batch = 16 ,lr0=0.01, dropout= 0.15, device = 0)
```

```python
# Loading the best performing model
Valid_model = YOLO(r'runs\detect\train9\weights\best.pt')

# Evaluating the model on the testset
metrics = Valid_model.val(split='val')
```

```python
import os
import cv2
from PIL import Image
from IPython.display import display

# --- Configuration ---
CONF_THRESHOLD = 0.25 # Minimum confidence score to draw a box
BOX_COLOR = (0, 255, 0) # Green color for bounding boxes (or choose based on class if needed)
BOX_THICKNESS = 2

# --- Get Image List ---
images_dir = "archive/brain-tumor/valid/images"
try:
    images = os.listdir(images_dir)
except FileNotFoundError:
    print(f"Error: Directory not found - {images_dir}")
    images = []

for i in range(min(9, len(images))):
    image_path = os.path.join(images_dir, images[i])
    result_predict = Valid_model.predict(source=image_path, imgsz=(640), iou=0.4, conf=CONF_THRESHOLD, verbose=False)
    results = result_predict[0]
    img_to_draw_on = results.orig_img.copy()
    if results.boxes is not None:
        boxes = results.boxes.cpu().numpy()

        for box_data in boxes:
            coords = box_data.xyxy[0].astype(int) # [xmin, ymin, xmax, ymax]
            xmin, ymin, xmax, ymax = coords
            cv2.rectangle(img_to_draw_on, (xmin, ymin), (xmax, ymax), BOX_COLOR, BOX_THICKNESS)


    plot_rgb = cv2.cvtColor(img_to_draw_on, cv2.COLOR_BGR2RGB)
    display(Image.fromarray(plot_rgb))
```
